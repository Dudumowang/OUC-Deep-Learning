```python
import torch
```

```python
x = torch.tensor([1,2,3,4,5,6])
print(x)
```

```python
# 2*3的值为1
x = torch.ones(2,3)
print(x)
```

```python
x = torch.ones(2,3,4)
print(x)
```

```python
x = torch.empty(5,3)
print(x)
```

```python
# 5*3的随机
x = torch.rand(5,3)
print(x)
```

```python
x = torch.zeros(5,3,dtype = torch.long)
print(x)
```

```python
y = x.new_ones(5,3)
print(y)
```

```python
# 形如x的
z = torch.randn_like(x,dtype=torch.float)
print(z)
```

```python
print(m.numel())
```

```python
print(m[0][2])
```

```python
print(m[:,1])
```

```python
print(m[0,:])
```

```python
from matplotlib import pyplot as plt
# randn生成均值为0，方差为1的随机数
plt.hist(torch.randn(1000).numpy(),100);
```
